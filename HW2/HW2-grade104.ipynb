### Question 2 - Independence and conditional independence

#### 2.A
Can we define a three dimensional multivariate gaussian that is pairwise-independent but not collectivley independent?
If your answer is yes, provide an example.
If your answer is no, provide a proof.

##### Answer

**NO** - A three dimensional multivariate gaussian that is pairwise-independent MUST be collectivley dependent.

<u>Proof</u>:<br>
If the 3 Gaussians are pairwise-independent, their covariance-matrix is diagonal (as seen in lecture 4):
$$\begin{bmatrix}
\sigma_1^2 & 0 & 0 \\
0 &  \sigma_2^2 & 0 \\
0 &  0 & \sigma_3^2
\end{bmatrix}$$

We denote this dist. by a vector $\vec{x} = [X=x, Y=y, Z=z]$

And, as seen in class, a multivariate normal dist. is defined by its PDF:
$$p(\vec{x}) = \frac1{\sqrt{(2\pi)^d \cdot Det\Sigma}}\exp(-\frac12 \cdot \langle \vec{x} - \vec{\mu}, \Sigma^{-1} (\vec{x} - \vec{\mu})\rangle)$$
(d = 3 --> number of dimensions and $\Sigma$ is the covariance matrix)

Given that the covariance matrix is diagonal (positive and semidefinite), when we plug it in the equation above we get:<br>

$$p(\vec{x}) = \frac1{\sqrt{(2\pi)^3 \cdot Det
\begin{bmatrix}
\sigma_1^2 & 0 & 0 \\
0 &  \sigma_2^2 & 0 \\
0 &  0 & \sigma_3^2
\end{bmatrix}
}}
\exp(-\frac12 
\begin{bmatrix}
x-\mu_x \\
y-\mu_y \\
z-\mu_z
\end{bmatrix} 
\begin{bmatrix}
\sigma_1^2 & 0 & 0 \\
0 &  \sigma_2^2 & 0 \\
0 &  0 & \sigma_3^2
\end{bmatrix}
^{-1}
\begin{bmatrix}
x-\mu_x \\
y-\mu_y \\
z-\mu_z
\end{bmatrix})$$

After applying some algebra as seen in class we get conclude that

$$P(X=x, Y=y, Z=z) = P(X=x)*P(Y=y)*P(Z=z)$$

which means that X, Y and Z are **collectively independent** by definition of collective independence.

#### 2.B
Let $\ X, Y \ $ and $Z$  be discrete random variables with $\ n, m \ $ and $k$ possible outcomes respectivley.

How many parameters define the joint distribution of $\ X, Y \ $ and $Z$?

##### Answer

**(n * m * k) - 1**<br>
-1 because all probabilities sum up to 1, and if we have (n * m * k) - 1 options, we can simply calculate the last prob. (1 - sum of all previous prob.s)

#### 2.C
For the same random variables from the previous section, how many parameters define the joint distribution of $\ X, Y \ $ and $Z$ if we now know that they are independent?

##### Answer

Independent means: $P(X=x, Y=y, Z=z) = P(X=x)*P(Y=y)*P(Z=z)$<br>
It's enough to look at each of the dist.s alone, because they are independent of one another.<br> 
**n + m + k - 3**<br>
Same as in section B, each dist. is independent, so all its prob.s sum up to one: (n-1) + (m-1) + (k-1)

#### 2.D
For the same random variables from the previous section, how many parameters define the joint distribution of $\ X, Y \ $ and $Z$ if we now know that $X$ and $Y$ are conditionaly independent given $Z$?

##### Answer

$P(X=x, Y=y, Z=z) = P(X|Z) * P(Y|Z) * P(Z)$<br>
For Z, we simply need (k-1), as in section C.<br>
For P(X|Z) it's (n-1)*k -> For each k, there are n options such that they all sum to 1 (for every k), so we can deduce the last prob. by calculating 1-sum(n-1). In total, there are k such scenarios.<br>
For P(Y|Z) it's (m-1)*k -> as above<br>
Overall, there are (n-1)*k*(m-1)*k*(k-1) = **(n-1)(m-1)(k-1)k^2** possible outcomes.

#### 2.E
For the random variables from the previous section, let $n=m=2$ and $k=2$. Give and example for a joint distribution of $\ X, Y \ $ and $Z$ where $X$ and $Y$ are conditionaly independent given $Z$?

##### Answer

We assume X and Y are dependent in general, such that given Z they become independent.<br>
Let **X**: First result of a coin (fair/unfair) chosen by Z is HEADS<br>
**Y**: Second result of a coin (fair/unfair) chosen by Z is HEADS<br>
**Z**: Choosing a fair coin out of 2 coins<br>
(0.5 for choosing each coin. The unfair coin, returns only HEADS)<br>

Z: fair coin<br>
$P(X|Z) = P(Y|Z) = 0.5$<br>
$P(X|Z)*P(Y|Z) = 0.25$<br>

Z: unfair coin<br>
$P(X|Z) = P(Y|Z) = 1$<br>
$P(X|Z)*P(Y|Z) = 1$<br>


$P(X) = 0.5 * 0.5 + 0.5*1 = 0.75$<br>
$P(Y) = 0.5 * 0.5 + 0.5*1 = 0.75$<br>
$P(X,Y) = 0.5 * 0.5 * 0.5 + 0.5 * 1 * 1 = 0.625$<br>
(<u>first scenario</u>: Z chooses a fair coin, X,Y are 0.5, <u>second scenario</u>: Z chooses an unfair coin, X,Y are 1)<br>
$P(X,Y) = 0.625 \neq P(X) * P(Y) = 9/16$<br>
So X and Y are dependent.<br>

Now we show that:
If Z was the fair coin:  $P(X,Y|Z) = P(X|Z) * P(Y|Z) = 0.25$ (two tosses of a fair coin)<br>
If Z was the unfair coin:  $P(X,Y|Z) = P(X|Z) * P(Y|Z) = 1$ (if the unfair coin s always HEADS, X and Y are always HEADS)<br>

So, for each of Z outcomes, $P(X,Y|Z) = P(X|Z) * P(Y|Z)$

### Question 3 - Gaussian mixtures – parameter estimation and generation 

Consider the data provided in GMD_2021.csv
Assume that the data comes from a Gaussian mixture distribution (GMD) with $k=3\ \ $. Furthermore, assume that $\mu_{1}=4, \mu_{2}=9, \sigma_{1}=\sigma_{2}=0.5, \sigma_3=1.5$ and $w_2=0.25$.

Read the data and answer the following questions.

#### 3.A
Provide an estimate for the other parameters of the distribution in two different ways.

##### Answer

There are (15+177+16) =~ 208 instances around the centered normal distribuiton and (59+206+175+50+2)~=492 around the rightmost distribution, so we can assume that $w_3 = 2.5*w_2 = 0.625$<br>
Since $\sum_{i=1}^3 w_i = 1$, we say that $w_1=1-0.25-0.625=0.125$

##### Another option to estimate the missing parameters is to use the **EM (Expectation Maximization)** algorithm, as learned in class:

We initialize the missing $\mu_3$ value based on the inferences from the histogram part above, but we could also initialize it randomly. 

```python
mu_1 = 4 # given
mu_2 = 9 # given
mu_3 = 15 # from visualizing the histogram

# given:
sigma_1 = 0.5
sigma_2 = 0.5
sigma_3 = 1.5

# ws are 0.5 (for starts)
w_1 = w_2 = w_3 = 0.33

df_1 = pd.DataFrame([[mu_1, sigma_1, w_1]], columns=['mean', 'sigma', 'w'])
df_2 = pd.DataFrame([[mu_2, sigma_2, w_2]], columns=['mean', 'sigma', 'w'])
df_3 = pd.DataFrame([[mu_3, sigma_3, w_3]], columns=['mean', 'sigma', 'w'])


def compute_responsibilities(data):
    p_1 = df_1['w'].values * norm.pdf(data, loc=df_1['mean'], scale=df_1['sigma'])
    p_2 = df_2['w'].values * norm.pdf(data, loc=df_2['mean'], scale=df_2['sigma'])
    p_3 = df_3['w'].values * norm.pdf(data, loc=df_3['mean'], scale=df_3['sigma'])
    
    denominator = (p_1 + p_2 + p_3)

    r_1 = p_1 / denominator
    r_2 = p_2 / denominator
    r_3 = p_3 / denominator
    
    return r_1, r_2, r_3


def compute_new_assignments(r_1, r_2, r_3, number_of_records):
    new_w_1 = r_1.sum() / number_of_records
    new_w_2 = r_2.sum() / number_of_records
    new_w_3 = r_3.sum() / number_of_records
    new_mu_3 = (r_3 * data).sum().values / (new_w_3 * number_of_records)

    return new_w_1, new_w_2, new_w_3, new_mu_3

################################

MAXIMUM_ITERATIONS = 1000
EPSILON = 0.00001
data = df_gmd
number_of_records = data.shape[0]

for i in range(MAXIMUM_ITERATIONS):
    r_1, r_2, r_3 = compute_responsibilities(data)
    new_w_1, new_w_2, new_w_3, new_mu_3 = compute_new_assignments(r_1, r_2, r_3, number_of_records)

    # stop condition
    if np.isclose(new_w_1, df_1['w'], atol=EPSILON) and \
            np.isclose(new_w_2, df_2['w'], atol=EPSILON) and \
            np.isclose(new_w_3, df_3['w'], atol=EPSILON) and \
            np.isclose(new_mu_3, df_3['mean'], atol=EPSILON):
        print(f'Convergence reached after {i} iterations')
        break
    
    df_1['w'] = new_w_1
    df_2['w'] = new_w_2
    df_3['w'] = new_w_3
    df_3['mean'] = new_mu_3

df_result = pd.concat([df_1, df_2, df_3], axis=0)
df_result.index = ['leftmost dist.', 'centered dist.', 'rightmost dist.']
df_result
```

We see that the estimates that the EM algorithm provides us are **VERY** similar to the ones we predicted above!!!<br>
It means that we did a good job with the visulaiztion and the counting of instances per distribuiton above.
That's why it took us only 3 iterations until reached convergence :)<br>
(The w's sum up to 1, so we can be assured it worked properly)

### Question 5 - Coupon collector

Let $T_{N}$ denote the waiting time for full single coupon collection with N different equiprobable coupon types

```python
# https://en.wikipedia.org/wiki/Coupon_collector%27s_problem#Calculating_the_expectation

from collections import defaultdict
from scipy import stats

class CouponCollector:
    def __init__(self, n):
        self.n = n
    
    def mean(self):
        return self.n * sum(1 / d for d in range(1, self.n + 1))
    
    def var(self):
        res = 0.0
        for i in range(1, self.n + 1):
            pi = (self.n - i + 1) / self.n
            res += (1 - pi) / (pi ** 2)
        return res
    
    @staticmethod
    def Conv2(P1, P2):
        if not P1:
            return P2

        Q = defaultdict(float)

        for x, px in P1.items():
            for y, py in P2.items():
                Q[x + y] += px * py

        return dict(Q) 
    
    def q5c(self, th):
        res_dist = {}
        
        for i in range(1, self.n + 1):
            xi = stats.geom((self.n - i + 1) / self.n)
            xi_dist = {j: xi.pmf(j) for j in range(1, th + 1)}
            res_dist = self.Conv2(res_dist, xi_dist)
            
        return 1 - sum(v for k, v in res_dist.items() if k <= th)
```
